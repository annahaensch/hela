## Probabalistic Model

### Background on Hidden Markov Models

A Markov Model is a model for a system that evolves over time that uses the concept of a Markov Chain. A Markov Chain is a sequence of states of the system, where the transition from each state to the next is determined using a transition model. Most of the time, this transition model is stochastic, which means that the next state in the sequence is sampled from a probability distribution which is somehow defined using the current state.

Now, if we can directly observe the state of the system in the training data, then this is sufficient to learn the tranistion model. However, in many applications like this problem, the state of the system is not directly observable (our training data does not include explicit labeling of the state of the industrial machines). Instead, we can observe some other quantity which is produces according to the state of the system to what is called the emission model. In our case, we can observe the sensor measurements, whose behavior is determined by the state of the system through the emission model. This kind of model is called a Hidden Markov Model (HMM). Often the state of the system is referred to as the "hidden state" or "latent state", and the observations that we can see are the "emissions".

### AutoRegressive Hidden Markov Models and Implementation for this Problem

We model our system as a Hidden Markov Model, with a latent state that can be one of three possible states: "Normal", "Faulty", or "Failed". We assume that the transitions between these states only happen in one direction, and that a system cannot directly transition from "Normal" to "Failed", which reduces the transition matrix down to two probablities that need to be inferred.

<img src="state-diagram.png" alt="drawing" width="400"/>

We also will see soon that the emissions are highly correlated with one another, which we can take advantage of by including a dependency of the emission on the previous emission. This kind of HMM is known as an AutoRegressive Hidden Markov Model, or arHMM. In the diagram below, $z$ is the latent state and $x$ is the emission. The arrows denote the flow of information, in other words, they define the dependency of one object on another. The top row of horizontal arrows are the transition model, and the vertical arrows are the traditional emission model. The bottom row of horizontal arrows represent the part of the emission model that incorporates the previous emission into determining the current emission.

<img src="arHMM.png" alt="drawing" width="500"/>

Our emission model will assume that the sensor measurements for each of the four sensor channels is normally distributed, with the mean of the distribution being the previous sensor measurement. The variances of the distributions, one for each channel for each of the three states, are parameters of the model that will be infered using Stochastic Variational Inference using the Pyro package.

## Data Exploration and Preparation

First, we import all the required packages that we'll need. The biggest one is the Pyro package, which is a probabilistic programming language for Python, using PyTorch on the backend. We'll be using Pyro to infer the parameters of the model using Stochastic Variational Inference (SVI).

```python
import glob, os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import pyro.distributions as dist
from pyro.distributions import constraints
from pyro.infer.autoguide import AutoMultivariateNormal, init_to_median
from pyro.infer import SVI, TraceEnum_ELBO, Predictive, infer_discrete
from pyro import poutine
import pyro
import torch
from tqdm.notebook import tqdm
```

We load in the data from the CSV files and convert the data to PyTorch tensors to prepare it for use with Pyro.

```python
# List of data files
data_files = glob.glob("exampleco_data/machine_*.csv")

# Read in each data file using Pandas and extract the data as a Numpy array
data = [pd.read_csv(f, index_col=0).to_numpy() for f in data_files]

# Convert the data into PyTorch tensors
data = torch.dstack([torch.tensor(d, dtype=torch.float32) for d in data])

# Rearrange the indices of the combined data
data = torch.transpose(data.T, 1, 2)
```

Now we'll split the data for training and validation. Half of the machines will be in the training set and the other half are in the validation set.

```python
# Split the data
data_train = data[:10, ...]
data_valid = data[10:, ...]
```

Let's take a quick peak at one of the sensor channels.

```python
# Plot the raw data
plt.figure(figsize=(20,5))
plt.plot(data_train[2, :, 0])
plt.xlim(1200, 2000)
```

We can see that autocorrelation which we mentioned earlier, as well as the outliers of the dataset, which we are going to now remove. We could have chosen to model the outlier generation process itself in the emission model that we're using, which would remove the need for preprocessing the data. However this complicates the model and its inference, and since the outliers in the data are straightforward to identify, we will just remove them now before performing inference. We will copy the original data and save it for making plots later.

We remove the outliers by replacing all points that are more than 3 standard deviations from the mean (which is we assume is zero) with the previous data point. We fit that standard deviation on the training data, and use it to also remove the outliers on the validation data.

```python
# Clone the original data before removing outliers
data_train_orig = torch.clone(data_train)
data_valid_orig = torch.clone(data_valid)

def remove_outliers(data, std, n_std=3):
    """Function to remove outliers by replacing with previous data point"""
    for i in tqdm(range(data.shape[0])):
        for j in range(data.shape[1]):
            for k in range(data.shape[2]):
                if torch.abs(data[i,j,k]) > n_std * std[k]:
                    if j == 0:
                        data[i,j,k] = 0.
                    else:
                        data[i,j,k] = data[i,j-1,k]

# Perform outlier removal
std = torch.std(data_train, dim=(0,1))
remove_outliers(data_train, std)
remove_outliers(data_valid, std)
```

Let's look at the same data with the ouliers removed.

```python
# Plot the cleaned data
plt.figure(figsize=(20,5))
plt.plot(data_train[2, :, 0])
plt.xlim(1200, 2000)
```

Let's also take a closer look at that autocorrelation in the sensor data. We make a scatter plot of the current sensor value versus the next sensor value. We see that the data from the normal operating mode create the "ellipse" shape along the diagonal, which quantifies how autocorrelated the nominal sensor data is. The scattering of points outside this ellipse represents the sensor during the faulty state, as the variance in the sensor differences between time points is much larger. This is the fundamental feature of the data that we are hoping to exploit to identify the transition point from normal to faulty.

```python
# Plot scatterplots for autocorrelation between sensor measurements
fig, axs = plt.subplots(nrows=2, ncols=2, figsize=(10, 10))
axs = axs.flatten()
for i in range(data_train.shape[2]):
    axs[i].scatter(data_train[:, :-1, i], data_train[:, 1:, i], s=2)
    axs[i].set_xlabel(f"Sensor {i} at Time $t$")
    axs[i].set_ylabel(f"Sensor {i} at Time $t+1$");
```

The third sensor does not exhibit this feature however, as the variance in the faulty state is similar to the variance in the normal state. For this reason we will ignore this sensor channel and only consider the other three channels.

```python
# Remove the third sensor channel from the training and validation data
data_train = data_train[..., [0, 1, 3]]
data_valid = data_valid[..., [0, 1, 3]]
data_train_orig = data_train_orig[..., [0, 1, 3]]
data_valid_orig = data_valid_orig[..., [0, 1, 3]]
```

## Inference of Model Parameters

Now we implement our probabalistic model in Pyro. We only have two "free" probabilities in the transition matrix, which we place "uniform" Beta distribution priors on.

For the variances of the emission model, we place Half-Normal distribution priors on them, and we choose the shape of these priors to reflect our prior knowledge of the sensor variances. For example, the variance in the sensor measurements of the failed state is very small, as the values are all close to zero.

```python
def model(data):
    """HMM implemented in Pyro"""
    # Get the dimensions of the data
    n_seq, T, dim_obs = data.shape
    
    # Define the transition probabilities and their prior distributions
    p00 = pyro.sample("p00", dist.Beta(1., 1.))
    p11 = pyro.sample("p11", dist.Beta(1., 1.))
    p = torch.tensor([[p00, 1. - p00, 0.], 
                      [0., p11, 1. - p11], 
                      [0., 0., 1.]])
    
    # Define the variances for each state's emissions and their prior distributions
    σ0 = pyro.sample("σ0", dist.HalfNormal(10.).expand([dim_obs]).to_event(1))
    σ1 = pyro.sample("σ1", dist.HalfNormal(20.).expand([dim_obs]).to_event(1))
    σ2 = pyro.sample("σ2", dist.HalfNormal(0.1).expand([dim_obs]).to_event(1))
    σ = torch.vstack([σ0, σ1, σ2])
    
    # Define the plate context for the dimension of the emission
    obs_plate = pyro.plate("obs", dim_obs, dim=-1)
    
    # Process each sequence, or each individual machine's data
    for i in pyro.plate("seq", n_seq):
        # Initialize the state as normal
        z = torch.tensor([0])
        
        # Initialize the first emission
        x = data[i, 0, :]
        
        # Traverse the sequence of states
        for t in pyro.markov(range(1, T)):
            # Sample the next state using the transition probabilities
            z = pyro.sample(f"z_{i}_{t}", dist.Categorical(p[z]), infer={"enumerate": "parallel"})

            # Sample emissions and condition on the dataset
            with obs_plate:
                x = pyro.sample(f"x_{i}_{t}", dist.Normal(x, σ[z.squeeze(-1)]), obs=data[i, t, :])
```

Now we define a guide to perform the SVI process, which is a parametric distribution which we optimize to be close to the posterior distribution. We choose a multivariate normal distribution for computational tractibility and ease of implementation. We also initialize the Evidence Lower Bound (ELBO) as the loss function for SVI, as well as the Adam optimizer.

```python
# Initialize the guide for SVI
guide = AutoMultivariateNormal(poutine.block(model, expose=["p00", "p11", "σ0", "σ1", "σ2"]),
                               init_loc_fn=init_to_median)

# Initialize the Adam optimizer
optim = pyro.optim.Adam({"lr": 0.01})

# Initialize the ELBO loss function
elbo = TraceEnum_ELBO(max_plate_nesting=1, strict_enumeration_warning=True)

# Initialize the SVI process
svi = SVI(model, guide, optim, elbo)
```

Now we perform the SVI process in a manner similar to how neural networks are trained using stocahstic gradient descent. We use minibatches of 2 sequences per SVI step, and perform 100 epochs.

```python
# Clear the previous parameter store in Pyro
pyro.clear_param_store()

# Define the batch size
batch_size = 2

# Initialize an array to store the loss history
loss = []
```

```python
# Number of Epochs to run for
n_epochs = 100

# Perform the epochs
for i in tqdm(range(n_epochs)):
    # Initialize a loss for the epoch
    l = 0
    
    # Get a random permutation of the training data
    idx = torch.randperm(data_train.shape[0])
    
    # Split the training data into minibatches
    for batch_idx in torch.split(idx, batch_size):
        # Perform the SVI step and save the 
        l += svi.step(data_train[batch_idx, ...])
        
    # Append the total loss of the epoch to the history
    loss.append(l)
```

We plot the loss history to verify that the SVI converges to a minimum.

```python
# Plot the loss history
plt.figure()
plt.plot(loss)
#plt.yscale("symlog")
plt.xlabel("Epoch")
plt.ylabel("ELBO");
```

## Inference of State Trajectories

Now that we've inferred the parameters of the model, we can infer the latent states of the validation data sets. We utilize the Viterbi-like inference of the Pyro's `infer_discrete` function to write a utility to infer the latent state trajectory using the inference results.

```python
@infer_discrete(first_available_dim=-2, temperature=1)
def viterbi_decoder(data, params):
    """Infers the latent states of a sequence using parameters found in the inference step"""
    # Get the shape of the data
    T, dim_obs = data.shape
        
    # Define a plate context for the data dimensions
    obs_plate = pyro.plate("obs", dim_obs)
    
    # Get the transition probabilities
    p00 = params["p00"]
    p11 = params["p11"]
    p = torch.tensor([[p00, 1. - p00, 0.], 
                      [0., p11, 1. - p11], 
                      [0., 0., 1.]])
    
    # Get the emission variances for each state
    σ = torch.vstack([params["σ0"], params["σ1"], params["σ2"]])

    # Initialize the first latent state to zero
    z = [torch.tensor(0)]
    
    # Initialize the mean of the first emission as the first emission
    x = data[0, :]
    
    # Traverse the state sequence
    for t in pyro.markov(range(1, T)):
        # Sample the next state using the transition probabilities
        z.append(pyro.sample(f"z_{t}", dist.Categorical(p[z[-1]]), infer={"enumerate": "parallel"}))
        
        # Sample the emissions and condition on the data
        with obs_plate:
            x = pyro.sample(f"x_{t}", dist.Normal(x, σ[z[-1].squeeze(-1)]), obs=data[t, :])
    
    # Return the state trajectory
    return z
```

Now we write a utility function to infer the latent states by repeatedly using the median of the posterior to infer the state trajectory. For each sample, we infer an entire possible state trajectory that agrees with the observed emissions, and then extract the time at which the system transitions from the "normal" state to the "faulty" state. We then plots a kernel density estimate of this transition time on top of one of the sensor channels to verify if the inferred transition lines up with the data.

```python
def predict(n_sample, data_mach, data_mach_orig, data_train, xlim):
    """Infer state trajectories by sampling from the posterior parameter distribution and plot the results"""
    # Initialize a tensor to hold the results of the trajectory inferals
    z = torch.zeros((n_sample, data_mach.shape[0]))
    
    # Initialize a tensor to hold the fault points for each sample
    fault_pt = torch.zeros(n_sample)
    
    # Sample the posterior multiple times and infer the trajectory
    for i in range(n_sample):
        # Use the inference utility to infer the state trajectory after sampling from the posterior
        z[i, :] = torch.stack(viterbi_decoder(data_mach, guide.median()))
        
        # Find the time at which the state transitions to faulty
        fault_pt[i] = torch.sum(z[i, :] == 0)
                
    # Initialie a figure
    fig = plt.figure(figsize=(20,5))
    ax1 = plt.gca()
    
    # Plot a KDE of the fault point on top of the sensor data
    ax2 = plt.gca().twinx()
    sns.kdeplot(fault_pt.numpy(), ax=ax2, fill=True, color="tab:red")
    sns.rugplot(fault_pt.numpy(), ax=ax2, color="tab:red")
    
    # Plot the original sensor data for the first availabe channel
    ax1.plot(range(data_mach_orig.shape[0]), data_mach_orig[:, 0], lw=1, color="tab:blue")
    
    # Plot settings
    ax1.set_xlim(*xlim)
    ax1.set_ylabel("Sensor History")
    ax2.set_ylabel("KDE of Fault Start")
```

We generate the plots for five machines in the validation set, whose data was witheld from the inference process until this point.

```python
# Number of posterior samples to take
n_sample = 100

# Limits for each plot, hand-determined for aesthetics
xlims = np.array([[200, 800],
                  [1500, 2000],
                  [0, 750],
                  [0, 500],
                  [0, 350]])

# Infer the state trajectory of 5 machines in the validation state
for mach in range(5):
    predict(n_sample, data_valid[mach, ...], data_valid_orig[mach, ...], data_train, xlims[mach, :])
```

Let's explain how to interpret these plots. The blue line-plot in the background is the history of one the sensor measurements, without the outlier removal. The red density function is the Kernel Density Estimate of the point in time where the state transitioned from "Normal" to "Faulty". We observe that this belief is reasonable, as the distribution overlaps with the region of sensor measurements that are abnormal and precede system failure. The intervals are not able to capture the exact start of the faulty stage, which is a limitation of the current HMM implementation. A better model would be able to identify the start of the fault more closer to the actual start of the fault, but our model was able to identify it before the state transitions from faulty to failed. An exception is the special case of the last figure, where the model identifies the transition too late. This is most likely due to the fact that at least one of the sensor channels does not exhibit any of the "faulty" behavior.

# Outlook and Potential Improvements

The first potential improvement to the model is to include the outlier generation process in the emission model. An emission model that samples normally from the previous emission, but also has a probability to return a measurement from some distribution of outliers, would remove the need to preprocess the data.

The second potential improvement would be to move to a more sophisticated HMM architecture. A deep HMM, which is an HMM that has a continous state space with neural networks for the transition and emission models could better capture the cyclical nature of the sensor readings. A logical extension of this model is a Recurrent Neural Network, which can take into account the dependencies of the state on more than one previous state, which is not possible with a basic HMM. These models are more sophisticated, and if they involve neural networks for modeling the transition and emission functions, then they will require much more data to train effectively. That being said, the Pyro package was built for infering the parameters of these kinds of models using Bayesian Inference. In fact, I learned about Deep HMMs from the Pyro documentation.

Potentially a more sophisticated model, or more training data could improve the model's performance in detecting the transition. Also, a labelled dataset would allow for better training, since the actual ground truth states could be used to condition the model and thus better learn the model parameters.
